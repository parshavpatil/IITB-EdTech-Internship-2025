```python
import os
import pandas as pd

stdata_path = 'C:\\Users\\parshav\\Desktop\\EdTech_Final\\data\\STData'     # adjust if needed
participant_folders = sorted(os.listdir(stdata_path))
feature_rows = []

for pid in participant_folders:
    folder = os.path.join(stdata_path, pid)
    row = {}

    # Helper function to extract features from full time series per file
    def extract_time_series_features(filename, prefix):
        path = os.path.join(folder, filename)
        if os.path.exists(path):
            df = pd.read_csv(path)
            for col in df.columns:
                if pd.api.types.is_numeric_dtype(df[col]):
                    # compute summary stats per entire time series column
                    row[f'{prefix}_{col}_mean'] = df[col].mean()
                    row[f'{prefix}_{col}_std'] = df[col].std()
                    row[f'{prefix}_{col}_max'] = df[col].max()
                    row[f'{prefix}_{col}_min'] = df[col].min()

    # Extract features from all modalities per participant
    extract_time_series_features(f'{pid}_EEG.csv', 'EEG')
    extract_time_series_features(f'{pid}_GSR.csv', 'GSR')
    extract_time_series_features(f'{pid}_EYE.csv', 'EYE')
    extract_time_series_features(f'{pid}_IVT.csv', 'IVT')
    extract_time_series_features(f'{pid}_TIVA.csv', 'TIVA')

    # For target labels, just take first row values as is (not time series)
    for target_file in [f'{pid}_PSY.csv', f'{pid}_ENG.csv']:
        tpath = os.path.join(folder, target_file)
        if os.path.exists(tpath):
            df_target = pd.read_csv(tpath)
            for col in df_target.columns:
                if col not in row:
                    row[f'TARGET_{col}'] = df_target[col].iloc[0]

    feature_rows.append(row)

# Create dataframe and save
feature_matrix = pd.DataFrame(feature_rows)
feature_matrix.to_csv('feature_matrix.csv', index=False)
print("Feature matrix saved.")
print(feature_matrix.head())
```
